```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../ressources/dataset.csv", sep=";")
```

```python
# Distribution de x_wins et is_draw
print("x_wins distribution:")
print(df["x_wins"].value_counts())

print("\nis_draw distribution:")
print(df["is_draw"].value_counts())
```

```python
df["x_wins"].value_counts().plot(kind="bar", title="Distribution x_wins")
plt.show()

df["is_draw"].value_counts().plot(kind="bar", title="Distribution is_draw")
plt.show()
```

```python
# Case la plus occupée par X dans les états de victoire
x_wins_df = df[df["x_wins"] == 1]

means = []

for i in range(9):
    mean_value = x_wins_df[f"c{i}_x"].mean()
    means.append(mean_value)
    print(f"Case {i} :", mean_value)

import numpy as np
grid = np.array(means).reshape(3,3)

import seaborn as sns
sns.heatmap(grid, annot=True)
plt.title("Occupation des cases par X (quand X gagne)")
plt.show()
```

```python
# Corrélation entre features
corr = df.corr()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap="coolwarm")
plt.title("Heatmap des corrélations")
plt.show()
```

```python
print("Corrélation avec x_wins")
print(df.corr()["x_wins"].sort_values(ascending=False))

print("\nCorrélation avec is_draw")
print(df.corr()["is_draw"].sort_values(ascending=False))
```
