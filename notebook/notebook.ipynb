```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../ressources/dataset.csv", sep=";")

df.head()
```

```python
df.shape
df.info()
```

```python
print("x_wins distribution:")
print(df["x_wins"].value_counts())

print("\nis_draw distribution:")
print(df["is_draw"].value_counts())
```

```python
df["x_wins"].value_counts().plot(kind="bar", title="Distribution x_wins")
plt.show()

df["is_draw"].value_counts().plot(kind="bar", title="Distribution is_draw")
plt.show()
```

```python
corr = df.corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
print("Case centrale (c4_x) vs victoire X:")
print(df.groupby("c4_x")["x_wins"].mean())
```

```python
for i in range(9):
    print(f"Case {i} (X):", df.groupby(f"c{i}_x")["x_wins"].mean())
```

```python
corr_xwins = df.corr()["x_wins"].sort_values(ascending=False)
corr_draw = df.corr()["is_draw"].sort_values(ascending=False)

print("Corrélation avec x_wins:")
print(corr_xwins)

print("\nCorrélation avec is_draw:")
print(corr_draw)
```
